```python
columns_to_convert = ['family_history_with_overweight', "FAVC", "SMOKE","SCC"]

# Iterate over each column in the dataset
for column in data_set.columns:
    # Check if the column contains 'yes' and 'no' values
    if 'yes' in data_set[column].values and 'no' in data_set[column].values:
        # Replace 'yes' with 1 and 'no' with 0
        data_set[column] = data_set[column].replace({'yes': 1, 'no': 0})

data_set
```

```python
# Convert 'Gender' column to categorical data type
data_set['Gender'] = data_set['Gender'].astype('category')
data_set['CAEC'] = data_set['CAEC'].astype('category')
data_set['CALC'] = data_set['CALC'].astype('category')
data_set['MTRANS'] = data_set['MTRANS'].astype('category')





# Convert 'Gender' column to numerical codes
data_set['Gender'] = data_set['Gender'].cat.codes
data_set['CAEC'] = data_set['CAEC'].cat.codes
data_set['CALC'] = data_set['CALC'].cat.codes
data_set['MTRANS'] = data_set['MTRANS'].cat.codes

data_set
```

```python
# Calculate BMI
data_set['BMI'] = data_set['Weight'] / (data_set['Height'] * data_set['Height'])

# Define the conditions for each BMI category
conditions = [
    (data_set['BMI'] < 18.5),
    (data_set['BMI'] >= 18.5) & (data_set['BMI'] <= 24.9),
    (data_set['BMI'] >= 25.0) & (data_set['BMI'] <= 29.9),
    (data_set['BMI'] >= 30.0) & (data_set['BMI'] <= 34.9),
    (data_set['BMI'] >= 35.0) & (data_set['BMI'] <= 39.9),
    (data_set['BMI'] >= 40.0)
]

# Define the corresponding BMI categories
categories = ['Underweight', 'Not Obesity', 'Overweight', 'Obesity I', 'Obesity II', 'Obesity III']

# Assign the corresponding category based on the conditions
data_set['Obezite'] = pd.Series(pd.cut(data_set['BMI'], bins=[0, 18.5, 24.9, 29.9, 34.9, 39.9, float('inf')], labels=categories))


# Convert the 'Height' column from float to integer
data_set['Height'] = (data_set['Height'] * 100).astype(int) / 100
data_set['Weight'] = (data_set['Weight'] * 100).astype(int) / 100
data_set['CH2O'] = data_set['CH2O'].astype(int)



# Convert the columns to integer
data_set['BMI'] = data_set['BMI'].astype(int)
data_set['Age'] = data_set['Age'].astype(int)
data_set['NCP'] = data_set['NCP'].astype(int)
data_set['FAF'] = data_set['FAF'].astype(int)
data_set['TUE'] = data_set['TUE'].astype(int)
data_set['FCVC'] = data_set['FCVC'].astype(int)



# Drop the column 'NObeyesdad'
data_set.drop(columns=['NObeyesdad'], inplace=True)

data_set
```
