```python
""" Gerekli kütüphaneler import ediliyor
Pandas kütüphanesini "import" komutuyla projemize ekliyoruz "pd" takma 
adıyla projemizin içinde kullanıyoruz ve verilerimizi işlemek ve analiz 
etmek için bu kütüphaneyi projemize dahil etmemiz gerekiyor."""


import pandas as pd
```

```python
"""data_set diye bir değişken oluşturuyoruz ve bu değişkene yukarda pd diye takma 
adıyla projemize dahil ettiğimiz pandas kütüphanesini kullanarak 
'data.csv' dosyasının içinde ki veri setini okuyoruz.
"""


data_set = pd.read_csv("data.csv")
data_set
```

```python
"""
Python DataFrame nesnesinde çağrılan bir yöntemdir. Bu yöntem, DataFrame'in bilgi veren bir özetini döndürür. 
Bu özet, DataFrame'in boyutu, sütunların adları, veri tipleri ve bellek kullanımı gibi bilgileri içerir.
Bu, DataFrame'in yapısı hakkında hızlı bir genel bakış sağlar ve veri hakkında daha fazla bilgi edinmeye yardımcı olabilir.
"""


data_set.info()
```

```python
"""
Bu kod bir veri kümesinden belirli bir sütunu (NObeyesdad olarak adlandırılan sütun) çıkarmak için kullanılır. 
data_set.drop("NObeyesdad", axis=1) ifadesi, data_set DataFrame'inden "NObeyesdad" sütununu çıkarır. 
axis=1 parametresi, bu işlemi sütun bazında yapılmasını sağlar. Yani, sütunlar boyunca işlem yapılacağını belirtir. 
Bu kod, DataFrame'deki "NObeyesdad" sütununu kalıcı olarak kaldırmaz, yalnızca işlem sırasında kullanılır. 
Eğer bu işlemi kalıcı hale getirmek istiyorsanız, inplace=True parametresini eklemeniz gerekir.

"""



data_set.drop("NObeyesdad",axis=1)
```

```python
"""
Bu kod parçası, belirli sütunlarda ( 'family_history_with_overweight', 'FAVC', 'SMOKE' ve 'SCC') bulunan 'yes' ve 'no' değerlerini 1 ve 0 olarak değiştirmek için bir döngü kullanır. 
Öncelikle, columns_to_convert listesinde belirtilen sütunlar üzerinde döner. Ardından, her bir sütun için 'yes' ve 'no' değerlerini içerip içermediğini kontrol eder. 
Eğer bir sütunda her iki değer de bulunuyorsa, 'yes' değerlerini 1 ile ve 'no' değerlerini 0 ile değiştirir. 
Bu, veri kümesindeki evet/hayır türündeki yanıtları sayısal bir formata dönüştürmeye yardımcı olur.


Sonuç olarak, döngü tamamlandığında, veri kümesindeki belirli sütunlardaki 'yes' ve 'no' değerleri, 1 ve 0 olarak değiştirilir. 
Bu dönüşüm, 'data_set' adlı DataFrame'in güncellenmiş bir versiyonunu oluşturur.


"""




columns_to_convert = ['family_history_with_overweight', "FAVC", "SMOKE","SCC"]


for column in data_set.columns:

    if 'yes' in data_set[column].values and 'no' in data_set[column].values:

        data_set[column] = data_set[column].replace({'yes': 1, 'no': 0})

data_set
```

```python

"""

Bu kod parçası, 'Gender', 'CAEC', 'CALC' ve 'MTRANS' sütunlarını kategorik veri türüne dönüştürür ve ardından bu sütunları sayısal kodlara dönüştürür.

İlk olarak, astype() yöntemi kullanılarak sütunlar kategorik veri türüne dönüştürülür.
Bu, belirli sütunların bir kategori olarak kabul edilmesini ve bu kategorilerin sınırlı bir sayıda önceden belirlenmiş değerler içereceğini belirtir.

Daha sonra, cat.codes kullanılarak kategorik sütunlar sayısal kodlara dönüştürülür. 
Bu, her kategori değerine bir tamsayı atar. Örneğin, eğer 'Gender' sütunu 'erkek' ve 'kadın' olmak üzere iki kategori içeriyorsa, 'erkek' değeri 0, 'kadın' değeri ise 1 olacaktır.

Sonuç olarak, DataFrame'deki 'Gender', 'CAEC', 'CALC' ve 'MTRANS' sütunları sayısal kodlara dönüştürülür. 

"""



data_set['Gender'] = data_set['Gender'].astype('category')
data_set['CAEC'] = data_set['CAEC'].astype('category')
data_set['CALC'] = data_set['CALC'].astype('category')
data_set['MTRANS'] = data_set['MTRANS'].astype('category')



data_set['Gender'] = data_set['Gender'].cat.codes
data_set['CAEC'] = data_set['CAEC'].cat.codes
data_set['CALC'] = data_set['CALC'].cat.codes
data_set['MTRANS'] = data_set['MTRANS'].cat.codes

data_set
```

```python


"""
BMI Hesaplama:

'Weight' (kilo) ve 'Height' (boy) sütunlarındaki değerler kullanılarak her bir satır için Vücut Kitle İndeksi (BMI) hesaplanır.
BMI hesaplaması, kişinin kilosunu boyunun karesine bölmek suretiyle yapılır.
Hesaplama sonucu elde edilen BMI değerleri, 'BMI' sütununa atanır.
Obezite Kategorileri İçin Koşullar ve Kategoriler:

BMI değerlerine dayanarak, kişinin obezite kategorisini belirlemek için koşullar belirlenir.
Belirlenen koşullara göre, kişinin hangi obezite kategorisine girdiğini belirlemek için kullanılacak kategoriler belirlenir.
Koşullar ve kategoriler, veri setindeki 'BMI' sütunundaki değerlere göre tanımlanır.
Obezite Sütunu Oluşturma:

Belirlenen koşullara ve kategorilere dayanarak, her bir kişinin obezite kategorisini içeren bir 'Obezite' sütunu oluşturulur.
'BMI' sütunundaki değerler, belirlenen koşullara göre kategorilere ayrılır ve bu kategorilere göre 'Obezite' sütununa atanır.


Veri Kümesi Sütunlarının Dönüşümü: , 'CH2O', 'Age', 'NCP', 'FAF', 'TUE' ve 'FCVC' sütunları tamsayı türüne dönüştürülür.


Bu kod, 'Height' sütunundaki değerleri santimetre cinsinden tutarlı bir formata dönüştürür. 
Öncelikle, mevcut yükseklik değerleri metreye cinsinden olduğunu varsayar. Kod, her bir yükseklik değerini 100 ile çarparak santimetre cinsinden bir değere dönüştürür. 
Daha sonra, bu değerleri 'int' (tamsayı) türüne dönüştürerek ondalık kısmı kaldırır. 
Son olarak, bu işlemin sonucunu tekrar 100'e böler, böylece orijinal ölçeklendirmeyi geri alır ve iki ondalık basamağa yuvarlar. 
Bu, yükseklik verilerinin birimleri üzerinde daha tutarlı bir işlem sağlar. Örneğin, 1.75 metrelik bir yükseklik değeri 175 santimetre olarak temsil edilecektir.




'BMI' sütunundaki değerlerin sayısını iki ondalık basamağa yuvarlar. 
Yani, her bir 'BMI' değeri iki ondalık basamağa kadar kesirli olarak yuvarlanır. 
Bu, veri kümesindeki BMI değerlerinin daha düzenli ve okunabilir olmasını sağlar. Örneğin, 22.456 olan bir BMI değeri 22.46 olarak yuvarlanacaktır.


'NObeyesdad' sütununu veri kümesinden kalıcı olarak kaldırır. .drop() yöntemi, belirtilen sütunu veri kümesinden kaldırmak için kullanılır. 
inplace=True parametresi, bu işlemi veri kümesine doğrudan uygular, yani kalıcı değişiklik yapar. 
'NObeyesdad' sütunu, veri kümesindeki gereksiz veya işlem yapılacak sütunlardan biri olarak kabul edilmiş olmalıdır. 
Bu kod, veri analizi veya modeleme sürecinde kullanılmayacak olan gereksiz bir sütunu temizler.


"""


data_set['BMI'] = data_set['Weight'] / (data_set['Height'] * data_set['Height'])

conditions = [
    (data_set['BMI'] < 18.5),
    (data_set['BMI'] >= 18.5) & (data_set['BMI'] <= 24.9),
    (data_set['BMI'] >= 25.0) & (data_set['BMI'] <= 29.9),
    (data_set['BMI'] >= 30.0) & (data_set['BMI'] <= 34.9),
    (data_set['BMI'] >= 35.0) & (data_set['BMI'] <= 39.9),
    (data_set['BMI'] >= 40.0)
]


categories = ['Underweight', 'Not Obesity', 'Overweight', 'Obesity I', 'Obesity II', 'Obesity III']


data_set['Obezite'] = pd.Series(pd.cut(data_set['BMI'], bins=[0, 18.5, 24.9, 29.9, 34.9, 39.9, float('inf')], labels=categories))



data_set['Height'] = (data_set['Height'] * 100).astype(int) / 100
data_set['Weight'] = (data_set['Weight'] * 100).astype(int) / 100


data_set['BMI'] = data_set['BMI'].round(2)


data_set['CH2O'] = data_set['CH2O'].astype(int)
data_set['Age'] = data_set['Age'].astype(int)
data_set['NCP'] = data_set['NCP'].astype(int)
data_set['FAF'] = data_set['FAF'].astype(int)
data_set['TUE'] = data_set['TUE'].astype(int)
data_set['FCVC'] = data_set['FCVC'].astype(int)





data_set.drop(columns=['NObeyesdad'], inplace=True)

data_set
```
